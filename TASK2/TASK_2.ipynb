### Implement an Al agent that plays the classic game of Tic-Tac-Toe against a human player. You can use algorithms like Minimax with or without Alpha-Beta Pruning to make the Al player unbeatable. This project will help you understand game theory and basic search algorithms

### Methodlogy Used:
#### The fundamental idea used here is the idea of back tracking using Minimax Alogithm
#### Since we need to let AI find the best possible area to place its choice, we need to:

If we consider user as 'MAX' and AI as 'MIN' then:

1) User will try to get maximum utility or profit or value
2) AI will try to fet least possible value between the choices (Which is favourable for it)


The time complexity for Minimax algorithm is O(b^d) where d is the branching factor (number of children) and d is the depth (Ply)

#### Case-1:Without Alpha-Beta Pruning

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np

# Size of the Tic Tac Toe board
N = int(input("Enter the size of the Tic Tac Toe board: "))

# Player and AI symbols
player, AI = 'X', 'O'

# Initialize the board
board = np.full((N, N), ' ')

# Function to evaluate the state of the board
def evaluate(board):
    # Check rows
    for row in range(0, N):
        if len(set(board[row])) == 1 and board[row][0] != ' ':
            if board[row][0] == AI:
                return +10
            elif board[row][0] == player:
                return -10

    # Check columns
    for col in range(0, N):
        if len(set(board[:, col])) == 1 and board[0][col] != ' ':
            if board[0][col] == AI:
                return +10
            elif board[0][col] == player:
                return -10

    # Check diagonals
    if len(set([board[i][i] for i in range(N)])) == 1 and board[0][0] != ' ':
        if board[0][0] == AI:
            return +10
        elif board[0][0] == player:
            return -10

    if len(set([board[i][N - i - 1] for i in range(N)])) == 1 and board[0][N - 1] != ' ':
        if board[0][N - 1] == AI:
            return +10
        elif board[0][N - 1] == player:
            return -10

    # If no one has won, return 0
    return 0

# Function to perform the Minimax algorithm
def minimax(board, depth, isMax):
    score = evaluate(board)

    # If AI has won the game, return his/her score
    if score == 10:
        return score

    # If player has won the game, return his/her score
    if score == -10:
        return score

    # If there are no more moves and no winner, return 0
    if np.all(board != ' '):
        return 0

    # If this is AI's move
    if isMax:
        best = -1000

        # Traverse all cells
        for i in range(N):
            for j in range(N):
                # Check cell is empty
                if board[i][j] == ' ':
                    # Make the move
                    board[i][j] = AI

                    # Call minimax recursively and choose the maximum value
                    best = max(best, minimax(board, depth + 1, not isMax))

                    # Undo the move
                    board[i][j] = ' '
        return best

    # If this is player's move
    else:
        best = 1000

        # Traverse all cells
        for i in range(N):
            for j in range(N):
                # Check cell is empty
                if board[i][j] == ' ':
                    # Make the move
                    board[i][j] = player

                    # Call minimax recursively and choose the minimum value
                    best = min(best, minimax(board, depth + 1, not isMax))

                    # Undo the move
                    board[i][j] = ' '
        return best

# Function to find the best move for AI
def findBestMove(board):
    bestVal = -1000
    bestMove = (-1, -1)

    # Traverse all cells, evaluate minimax function for all empty cells. And return the cell with optimal value.
    for i in range(N):
        for j in range(N):
            # Check cell is empty
            if board[i][j] == ' ':
                # Make the move
                board[i][j] = AI

                # Compute evaluation function for this move
                moveVal = minimax(board, 0, False)

                # Undo the move
                board[i][j] = ' '

                # If the value of the current move is more than the best value, then update best
                if moveVal > bestVal:
                    bestMove = (i, j)
                    bestVal = moveVal

    return bestMove

# Function to check if the game is over
def isGameOver(board):
    return evaluate(board) != 0 or np.all(board != ' ')

# Function to handle button click
def on_button_click(button):
    # Get button position
    i, j = button_positions[button]
    
    # Player's move
    if board[i][j] == ' ':
        board[i][j] = player
        button.description = player
        button.disabled = True
    
    if isGameOver(board):
        for b in buttons:
            b.disabled = True
        if evaluate(board) == 10:
            print("AI wins!")
        elif evaluate(board) == -10:
            print("Player wins!")
        else:
            print("It's a tie!")
        return
    
    # AI's move
    i, j = findBestMove(board)
    board[i][j] = AI
    buttons[i * N + j].description = AI
    buttons[i * N + j].disabled = True
    
    # Check if game is over
    if isGameOver(board):
        for b in buttons:
            b.disabled = True
        if evaluate(board) == 10:
            print("AI wins!")
        elif evaluate(board) == -10:
            print("Player wins!")
        else:
            print("It's a tie!")
        return

# Create buttons for the board
buttons = []
button_positions = {}
for i in range(N):
    for j in range(N):
        button = widgets.Button(description=' ')
        button.style.button_color = 'red'
        button.style.font_weight = 'bold'
        button.style.font_size = '20px'
        button.on_click(on_button_click)
        buttons.append(button)
        button_positions[button] = (i, j)

# Display the board
for i in range(N):
    display(widgets.HBox(buttons[i * N : (i + 1) * N]))
```

Setting the Depth parameter to 3 or 4 tells AI to see 3 or 4 moves ahead

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np

""""
+10 - AI
-10 - Player
"""
# Size of the Tic Tac Toe board
N = int(input("Enter the size of the Tic Tac Toe board: "))

# Player and AI symbols
player, AI = 'X', 'O'

# Initialize the board
board = np.full((N, N), ' ')

# Function to evaluate the state of the board
def evaluate(board):
    # Check rows
    for row in range(0, N):
        if board[row][0] == board[row][1] and board[row][1] == board[row][2]:
            if board[row][0] == AI:
                return +10
            elif board[row][0] == player:
                return -10

    # Check columns
    for col in range(0, N):
        if board[0][col] == board[1][col] and board[1][col] == board[2][col]:
            if board[0][col] == AI:
                return +10
            elif board[0][col] == player:
                return -10

    # Check diagonals
    if board[0][0] == board[1][1] and board[1][1] == board[2][2]:
        if board[0][0] == AI:
            return +10
        elif board[0][0] == player:
            return -10

    if board[0][2] == board[1][1] and board[1][1] == board[2][0]:
        if board[0][2] == AI:
            return +10
        elif board[0][2] == player:
            return -10

    # If no one has won, return 0
    return 0

#DEPTH PARAMETER TO BE UPDATED HERE

def minimax_algorithm(board, depth, isMax, alpha, beta):
    score = evaluate(board)

    # If AI has won the game, return score
    if score == 10:
        return score

    # If player has won the game, return score
    if score == -10:
        return score

    # If there are no more moves and no winner, return 0
    if np.all(board != ' ') or depth == 0:
        return 0

    # If AI's move
    if isMax:
        best = -1000

        # Traverse all cells
        for i in range(N):
            for j in range(N):
                # Check cell is empty
                if board[i][j] == ' ':
                    # Make the move
                    board[i][j] = AI

                    # Call minimax recursively and choose the maximum value
                    best = max(best, minimax_algorithm(board, depth - 1, not isMax, alpha, beta))

                    # Undo the move
                    board[i][j] = ' '

                    # Alpha Beta Pruning
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        return best
        return best

    # If player's move
    else:
        best = 1000

        # Traverse all cells
        for i in range(N):
            for j in range(N):
                # Check cell is empty
                if board[i][j] == ' ':
                    # Make the move
                    board[i][j] = player

                    # choose the minimum value
                    best = min(best, minimax_algorithm(board, depth - 1, not isMax, alpha, beta))

                    # Undo the move
                    board[i][j] = ' '

                    # Alpha Beta Pruning
                    beta = min(beta, best)
                    if beta <= alpha:
                        return best
        return best

def on_clear_button_click(button):
    # Clear the output
    clear_output(wait=True)
    
    # Reset the board
    global board
    board = np.full((N, N), ' ')
    
    # Reset the buttons
    for button in buttons:
        button.description = ' '
        button.disabled = False

# Create a clear button
clear_button = widgets.Button(description='Clear')
clear_button.on_click(on_clear_button_click)

# Function to find the best move for AI
def findBestMove(board):
    bestVal = -1000
    bestMove = (-1, -1)
    
    for i in range(N):
        for j in range(N):
            # Check cell is empty
            if board[i][j] == ' ':
                # Make the move
                board[i][j] = AI

                # Evaluation function for this move
                moveVal = minimax_algorithm(board, 3, False, -1000, 1000)  # Set depth to 3 to see 3 moves ahead

                # Undo the move
                board[i][j] = ' '

                # If the value of the current move is more than the best value, then update best
                if moveVal > bestVal:
                    bestMove = (i, j)
                    bestVal = moveVal

    return bestMove

# Function to check if the game is over
def isGameOver(board):
    return evaluate(board) != 0 or np.all(board != ' ')

# Function to handle button click
def on_button_click(button):
    i, j = button_positions[button]
    
    # Player's move
    if board[i][j] == ' ':
        board[i][j] = player
        button.description = player
        button.disabled = True
    
    if isGameOver(board):
        for b in buttons:
            b.disabled = True
        if evaluate(board) == 10:
            print("AI wins!")
        elif evaluate(board) == -10:
            print("Player wins!")
        else:
            print("It's a tie!")
        return
    
    # AI's move
    i, j = findBestMove(board)
    board[i][j] = AI
    buttons[i * N + j].description = AI
    buttons[i * N + j].disabled = True
    
    # Check if game is over
    if isGameOver(board):
        for b in buttons:
            b.disabled = True
        if evaluate(board) == 10:
            print(f"AI wins!")
        elif evaluate(board) == -10:
            print("Player wins!")
        else:
            print("It's a tie!")
        return

# Create buttons for the board
buttons = []
button_positions = {}
for i in range(N):
    for j in range(N):
        button = widgets.Button(description=' ')
        button.style.button_color = 'red'
        button.style.font_weight = 'bold'
        button.style.font_size = '20px'
        button.on_click(on_button_click)
        buttons.append(button)
        button_positions[button] = (i, j)

# Display the board
for i in range(N):
    display(widgets.HBox(buttons[i * N : (i + 1) * N]))
display(clear_button)
```
